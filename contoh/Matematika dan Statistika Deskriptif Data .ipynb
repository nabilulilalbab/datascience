### Mendapatkan nilai mean

```python
import numpy as np

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = np.mean(score)
```

### Mendapatkan nilai median

```python
import numpy as np

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = np.median(score)
```

### Mendapatkan nilai mode

```python
from scipy import stats

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = stats.mode(score)
```

### Mendapatkan nilai quartile dan interquartile range

```python
import numpy as np

data = [3, 4, 4, 5, 5, 6, 6, 7, 8, 8, 9, 10]

Q1 = np.median(data[:int(len(data)/2)])
Q3 = np.median(data[int(len(data)/2):])
IQR = Q3 - Q1
```

```python
Q1, Q3, IQR
```

### Mendapatkan nilai variance dan standard deviation

```python
import numpy as np

data= [86, 87, 88, 86, 87, 85, 86]

mean = np.mean(data)
var = np.var(data)
std = np.std(data)
```

```python
mean, var, std
```
