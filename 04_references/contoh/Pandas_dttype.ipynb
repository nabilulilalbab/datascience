### Membuat Series dari sebuah List Python

```python
import pandas as pd

li = [1, 7, 2]
se = pd.Series(li)

print(se)
```

### Mengakses elemen pada Series

```python
li = [1, 7, 2]
se = pd.Series(li)

print(se[0])
```

### Memberi label pada Series

```python
li = [1, 7, 2]
se = pd.Series(li, index = ['x', 'y', 'z'])

print(se['x'])
```

### Membuat Series dari Dictionary

```python
calories = {"day1": 420, "day2": 380, "day3": 390}
se = pd.Series(calories)

print(se)
```

```python
pd.merge(left=add_df, right=more_df, how='inner', on='ID')
```

```python
pd.merge(left=add_df, right=more_df, how='outer', on='ID')
```

```python
pd.merge(left=add_df, right=more_df, how='left', on='ID')
```

```python
pd.merge(left=add_df, right=more_df, how='right', on='ID')
```
